```python
# Problem 1: Smart Parking Lot Management System

def smart_parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    if parked > capacity:
        status = "Over Capacity Alert"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)
    print("Parking Status:", status)

# Example
smart_parking_system(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    eta = distance * 5

    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 8

    if weather == "Rainy":
        eta += 10
    elif weather == "Storm":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")

# Example
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def theatre_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", str(int(occupancy)) + "%")
    print("Show Status:", status)

# Example (150 booked seats)
theatre_occupancy(200, [1]*150)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", str(int(average)) + "%")
    print("Server Status:", status)

# Example
classify_server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def classroom_resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


# Example
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def event_registration_controller(capacity, registrations):
    confirmed = 0

    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1

    waitlisted = registrations - confirmed

    if registrations >= capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

# Example
event_registration_controller(100, 105)
```
