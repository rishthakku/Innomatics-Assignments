4. Explain the difference between stemming and lemmatization.

Stemming is a rule-based technique that removes word endings using simple heuristics, whereas lemmatization uses vocabulary and grammatical analysis to return the correct base form of a word.

The output of stemming may sometimes be incorrect or not a real word, while lemmatization always produces a valid and meaningful word.

For example, the word "running" might be reduced to "run" or even "runn" in stemming, whereas in lemmatization it is correctly converted to "run".

```python
# Task 2
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    text = re.sub(r"\d+", "", text)

    text = text.lower()

    text = re.sub(r'(.)\1{2,}', r'\1', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    clean_sentence = " ".join(filtered_tokens)

    return filtered_tokens, clean_sentence
```

```python
# Task 3
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "Innomatics is THE best course!!!",
    "Visit http://example.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)

    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)

    results.append(tokens)
```

```python
# Task 4
import numpy as np

analytics = []

for tokens in results:
    if len(tokens) == 0:
        analytics.append((0, 0, 0))
        continue

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens])

    analytics.append((total_tokens, unique_tokens, avg_length))

for i, (t, u, a) in enumerate(analytics):
    print(f"\nSentence {i+1}:")
    print("Total Tokens:", t)
    print("Unique Tokens:", u)
    print("Avg Token Length:", round(a, 2))
```

```python
# Task 5
from collections import Counter

all_tokens = [token for sentence in results for token in sentence]

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
# Task 6
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Example
pipeline_output = full_pipeline(test_sentences)
print(pipeline_output)
```

```python
# Task 7
def preprocess_text(text):
    try:
        if not isinstance(text, str):
            return [], ""

        if text.strip() == "":
            return [], ""

        text = re.sub(r"http\S+|www\S+|https\S+", "", text)
        text = re.sub(r"\S+@\S+", "", text)

        text = re.sub(r"\d+", "", text)

        text = text.lower()

        text = re.sub(r'(.)\1{2,}', r'\1', text)

        text = re.sub(r'[^a-z\s]', '', text)

        text = re.sub(r'\s+', ' ', text).strip()

        tokens = text.split()

        tokens = [
            word for word in tokens
            if len(word) > 2 or word in ["no", "not"]
        ]

        clean_sentence = " ".join(tokens)

        return tokens, clean_sentence

    except Exception as e:
        print(f"Error processing text: {e}")
        return [], ""
```

```python
edge_cases = [
    "",                     # Empty string
    "     ",                # Only spaces
    "😂😂😂😂",             # Only emojis
    "123456789",            # Only numbers
    None,                   # None input
    ["not", "a", "string"], # Invalid type
]

for case in edge_cases:
    print("\nInput:", case)
    tokens, sentence = preprocess_text(case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", sentence)
```
